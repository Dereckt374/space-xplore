```python
import requests
import pandas as pd
```

```python
IARU_adresses = {
'mission_coordinated' : 'https://iaru.amsat-uk.org/finished.php',
'request_pending' : 'https://iaru.amsat-uk.org/formal.php',
'mission_declined' : 'https://iaru.amsat-uk.org/declined.php',
'mission_other_info' : 'https://iaru.amsat-uk.org/informal.php',
}
```

```python
df = pd.DataFrame()
for i in IARU_adresses:
    dft = pd.read_html(IARU_adresses[i])[2].drop(0).drop([2,3,], axis=1).rename({0:'sat_name', 1:'sat_description'}, axis=1)
    dft['source'] = i
    df = pd.concat([df, dft])
df = df.reset_index()
```
