```python
def get_ST_query(link, auth) : 
    # URL pour l'authentification et la requête
    auth_url = 'https://www.space-track.org/ajaxauth/login'

    session = requests.Session()
    response = session.post(auth_url, data=auth_payload)
    if response.status_code == 200:
        print("Authentification réussie")
    else:
        print("Erreur lors de l'authentification")
        return(response.text)

    response = session.get(link)

    if response.status_code == 200:
        data = response.json()
        return(pd.DataFrame(data))
```
