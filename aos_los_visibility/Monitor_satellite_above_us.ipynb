```python
link = 'https://db.satnogs.org/transmitters'
df=pd.read_html(link)[0]
```

```python
df.drop('Unnamed: 0', axis = 1, inplace=True)
```

```python
df['Downlink'] = pd.to_numeric(df['Downlink'], errors='coerce')  ## Convertion des string en int, set to np.nan si impossible
df = df[df['Downlink'].notna()].astype({'Downlink': float})  ## Supprime toutes les lignes où la colonne est en np.nan
```

```python
freq_ENSO = df[(df['Satellite'].str.contains('ENSO')) & (df['Downlink'] > 100e6)]['Downlink'].to_list()[0]
print(f'fréquence de ENSO : {freq_ENSO:,} Hz')
```

```python
marge = 0.05
print(f'plage de fréquencre : {freq_ENSO * (1-marge):,} - {freq_ENSO * (1+marge):,}')
df_freq = df[((df['Downlink'] > freq_ENSO * (1-marge))) & (df['Downlink'] < (freq_ENSO * (1+marge))) & (df['Mode'] == 'GMSK') & (df['Baud'] == 2400)] #  (df['Status'] == 'active') 
```
